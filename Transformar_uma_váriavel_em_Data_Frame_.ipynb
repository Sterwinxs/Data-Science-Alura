```python
import pandas as pd
```

 Leitura de Arquivo CSV, atribuindo o próprio para uma Variável.

```python
tmdb= pd.read_csv("tmdb_5000_movies.csv")
```

Transformando uma váriavel em um Data Frame.

```python
contagem_de_lingua = tmdb.original_language.value_counts().to_frame().reset_index() # Transformamos as informações para um Data frame.
```

Podemos observar que a coluna 1 apareceu como o index, e o número de totais como original_language, iremos arrumar!

```python
contagem_de_lingua
```

```python
contagem_de_lingua.columns = ["original_language", "Total"]
contagem_de_lingua.head()
```
